#### 1 - Using a single join and a grouping, write a query that allows you to find the forename, surname, driverid and the total number of wins for each driver in the f1db schema.

I am not certain what wins in the results table is represented by. I have chosen to consistently use "rank = 1" to determine a win or not. 

```python
%%sql SELECT forename, surname, driverid, count(rank=1) wins
FROM drivers 
JOIN results using (driverid)
GROUP BY forename, surname, driverid
ORDER BY wins
```

#### 2 - Write a query that joins the tables results, constructors and drivers to show the amount of times a driver has driven a car from a constructor. In other words how many times one driver (for instance Schumacher) has driven a constructor (for instance Mercedes). Your table should have three fields: drivers.driverref, constructors.name and count(*) (showing the number of times the driver has driven the constructor).
##### 3 Hint: You can group on more than one value


```python
%%sql SELECT drivers.driverref, constructors.name, count(*) timesdriven
FROM results 
JOIN drivers using (driverid) 
JOIN constructors using (constructorid)
GROUP BY drivers.driverref, constructors.name
ORDER BY count(*)
```

#### 4 - Now we have found the most driven vehicles even for those who didn't finish the race. Extend the query from 2 by removing all tuples from the result where the status is not "Finished".

```python
%%sql SELECT drivers.driverref, constructors.name, count(*) timesdriven
FROM results 
JOIN drivers using (driverid) 
JOIN constructors using (constructorid)
WHERE statusid = 1
GROUP BY drivers.driverref, constructors.name
ORDER BY count(*)
```

#### 5: Create a third (and new) query that finds the amount of milliseconds spent in pitstop (see the pitstops relation) by each unique combinations of driverid and raceid.
##### 6 Hint: use the sum aggregate function to find the total amount of milliseconds
##### 7 Hint: i recommend making this into a view

Firstly creating view


```python
%%sql CREATE VIEW view_totalms AS
SELECT driverid, raceid, sum(milliseconds) totalmilliseconds
FROM pitstops
GROUP BY driverid, raceid
ORDER BY totalmilliseconds
```

Running view

```python
%sql SELECT * FROM view_totalms;
```

#### 8 - Time to put everything together. Use the query for 4 as a join subquery in query 3 (see slides on "a query within a query") to find the total amount of pitstop time, for each result in query 3

```python
%%sql CREATE VIEW view_totalpitstop AS
SELECT drivers.driverref, constructors.name, count(*) timesdriven, sum(totalmilliseconds) totalmilliseconds
FROM results 
JOIN drivers using (driverid) 
JOIN constructors using (constructorid)
JOIN view_totalms using (driverid, raceid)
WHERE statusid = 1
GROUP BY drivers.driverref, constructors.name
ORDER BY totalmilliseconds
```

Running view

```python
%sql SELECT * FROM view_totalpitstop;
```

#### 9 - As a last thing we want to find the driver with the least pistop time. However we want to control for the amount of races (see explanation below), so we need to include one last column: the average pitstop time per race. Your table should be sorted based on the ascending values of that average pitstop time.

##### 10 Explanation: If we just count the pitstop time, we only get the total time he spend in a workshop, no matter how many races he participated in. If a driver participated in 100 races, he would have more pitstop time than a driver participating in one race!
##### 11 Hint: Use the sum function to accumulate the total pistop time and divide that with the amount of races the driver has participated in

```python
%%sql CREATE VIEW view_totalpitstop_everyone AS
SELECT drivers.driverref, constructors.name, count(*) timesdriven, sum(totalmilliseconds) totalmilliseconds
FROM results 
JOIN drivers using (driverid) 
JOIN constructors using (constructorid)
JOIN view_totalms using (driverid, raceid)
GROUP BY drivers.driverref, constructors.name
ORDER BY totalmilliseconds
```

```python
%%sql SELECT driverref, races, sum(totalmilliseconds/races)
FROM view_totalpitstop_everyone
JOIN (
    SELECT driverref, count(driverid) races FROM results
    JOIN drivers USING (driverid)
    GROUP BY driverref
    ORDER BY count(driverid)
) temp USING (driverref)
GROUP BY driverref, races
ORDER BY sum(totalmilliseconds/races) 
LIMIT 1;
```
